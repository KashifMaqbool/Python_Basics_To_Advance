## 1.2 Importing datetime class from date time module

```python
from datetime import datetime

# Printing the current date
date_time = datetime.now()
print("Current Date :",date_time)

# Printing the current year, month and day individually
print("Year :",date_time.year)
print("Month :",date_time.month)
print("Day :",date_time.day)
```

## 1.3 Printing date and time using strftime() method

```python
# Printing current date

dt = datetime.now()

year = dt.strftime('%Y')
month = dt.strftime('%m')
day = dt.strftime('%d')

print("Year :", year)
print("Month :", month)
print("Day :", day)
```

```python
# Printing current Time

hours = dt.strftime('%H')
minutes = dt.strftime('%M')
seconds = dt.strftime('%S')

print("Hour :", hours)
print("Minutes :", minutes)
print("Seconds :", seconds)
```

```python
# Printing date and time together

dt1 = datetime.now()
date_and_time = dt1.strftime('%H:%M:%S   %d/%m/%y') 
print("date_and_time :", date_and_time)
```

```python
# Some formatted codes
dt2 = datetime.now()

# For days
days = dt2.strftime('%a, %A, %d') 
print("Days Versions :", days)

# For week
week = dt2.strftime('%w') 
print("Week Number :", week)

# For months
months = dt2.strftime('%b, %B, %m') 
print("Months Versions :", months)

# For year
year = dt2.strftime('%y, %Y') 
print("Year Versions :", year)
```
